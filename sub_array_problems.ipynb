# Longest subarry having sum == k
Given an array and a sum k, we need to print the length of the longest subarray that sums to k.

```python
#### in greedy way:
# 1. find all the sub arrays in N^2 time.
# 2. take the sum and compare with k if match
# 3. then take the length and compare with largest length till then.
# TC: O(N^3) N^2 for finding the subarrays, and O(N) for computing sum, finding out length.abs
# SC: O(1), no extra space is needed.


'''
Example 1:
Input Format: N = 3, k = 5, array[] = {2,3,5}
Result: 2
Explanation: The longest subarray with sum 5 is {2, 3}. And its length is 2.

Example 2:
Input Format: N = 5, k = 10, array[] = {2,3,5,1,9}
Result: 3
Explanation: The longest subarray with sum 10 is {2, 3, 5}. And its length is 3.

###To solve in O(N) time need to use two pointer approach.

1. create two pointers left and right and a variable sum.
2. if sum == k, the right-left + 1 (index is 0 based.)
3. if sum > k, the need to increase left pointer till the variable sum < k, if means for that index left we have reached the max subarry that we can consider now we move on.

Node: when adding to sum check if right < n, after adding increment right
'''

# arr = [2,3,5]
# k = 5

# arr = [2,3,5,1,9]
arr = [2,3,5]
k = 10

def longestSubarrayWithSumK(a: [int], k: int) -> int:

    left = 0
    right = 0
    n = len(arr)
    curr_sum = 0
    max_length = float("-inf")

    while right <= n:

        ####check left, dont want left to exceed right.
        while left <= right and  curr_sum > k:
            curr_sum -= arr[left]
            left += 1

        ###check if sum == k
        if curr_sum == k:
            max_length = max(max_length, right-left)

        if right < n:
            curr_sum += arr[right]

        right += 1

    return max_length

print(max_length)
```

## Longest Subarray == k, Solving using Prefix SUM

This uses a hash map that stores the prefix sum, for lookup.

Steps
1. iterate from 0 - n
2. keeping tack of sum by adding it to variable say curr_sum
3.1 if curr_sum == k then compare its length with the max_length still then.
3.2 else check if curr_sum  - k, is present in the hashmap, again calculate its length and compare with max_length
3.3  curr_sum is not present in hashmap add it to hashmap along with its index.

k = 5
curr_sum = 10

rem = curr_sum - k >> rem = 5

k = curr_sum - rem >> there exists a subarray with sum == k from index (rem) to index (curr_sum).


TC: O(N) or O(N*logN)

SC: O(N)

```python
def longestSubarrayWithSumK(a: [int], k: int) -> int:

    hashmap = {}
    n = len(a)
    curr_sum = 0
    max_length = float('-inf')

    for i in range(n):

        curr_sum += a[i]

        # print(curr_sum, i)

        if curr_sum == k:
            max_length = max(max_length, i+1)
        else:
            rem = curr_sum - k

            if rem in hashmap:
                max_length = max(max_length, i - hashmap[rem])

        if curr_sum not in hashmap:
            ###add the elements to hashmap
            hashmap[curr_sum] = i

    return max_length

a = [2,3,5]
k = 5
print(longestSubarrayWithSumK(a, k))

a = [2,3,5,1,9]
k = 10

print(longestSubarrayWithSumK(a, k))

###works for negative numbers as well.
a = [0,-2,2,2,3,5,-9,1,9]
k = 10

print(longestSubarrayWithSumK(a, k))
```

## Longest Consecutive Sequence in an Array
Problem Statement: You are given an array of ‘N’ integers. You need to find the length of the longest sequence which contains the consecutive elements.

Example 1:

Input: [100, 200, 1, 3, 2, 4]

Output: 4

Explanation: The longest consecutive subsequence is 1, 2, 3, and 4.

Input: [3, 8, 5, 7, 6]

Output: 4

Explanation: The longest consecutive subsequence is 5, 6, 7, and 8.

### Important things to note here.
1. we can reorder the sequence.

TC: O(N) + O(NlogN)

SC: O(1)

```python
'''
1. Sort the array : nlogn
2. if arr[i] > arr[i+1] then increment length.
'''

from typing import List

# arr = [100, 200, 1, 3, 2, 4]
arr = [3, 8, 5, 7, 6]

def longestSuccessiveElements(arr : List[int]) -> int:

    arr.sort()

    max_length = float('-inf')
    length = 1
    prev = arr[0]

    for i in arr:

        if prev + 1 == i:
            length += 1
        ###there can be two cases
        ###1. current element is duplicate of prev in such case we dont do anything.
        ###2. it is not a contiguious element.
        elif i != prev:
            length = 1

        max_length = max(max_length, length)

        prev = i

    return  max_length

arr = [3, 8, 5, 7, 6]

print(longestSuccessiveElements(arr))

arr = [1, 2, 5, 6,9,10,11,12,13,20]

print(longestSuccessiveElements(arr))
```

## Longest Consecutive Sequence in an Array Approach II

In this approach we are doing to use set. where searching will be O(1). Additionally can only search for elements who are the first element of a sequence or of element-x is not present in the array.




```python
def longestSuccessiveElements(arr : List[int]) -> int:

    search_space = set(arr) #TC and SC: O(N)

    max_length = float('-inf')

    for i in arr:

        if i - 1 not in search_space:

            length = 1
            x = i
            while x + 1 in search_space:
                length += 1
                x += 1

            max_length = max(max_length, length)

    return  max_length

arr = [3, 8, 5, 7, 6]

print(longestSuccessiveElements(arr))

arr = [1, 2, 5, 6,9,10,11,12,13,20]

print(longestSuccessiveElements(arr))

arr = [6, 3, 14, 18, 8, 18, 4, 3, 12, 1, 14, 7, 9, 22, 11, 28, 24, 22, 30, 13, 2, 19, 8, 8, 18, 14, 11, 13, 1]

print(longestSuccessiveElements(arr))
```

## Count the number of subarrays with given xor k

Example 1:

Input Format: A = [4, 2, 2, 6, 4] , k = 6

Result: 4

Explanation: The subarrays having XOR of their elements as 6 are  [4, 2], [4, 2, 2, 6, 4], [2, 2, 6], [6]

Example 2:

Input Format: A = [5, 6, 7, 8, 9], k = 5

Result: 2

Explanation: The subarrays having XOR of their elements as 5 are [5] and [5, 6, 7, 8, 9]

```python
from collections import defaultdict

def subarraysWithSumK(a: [int], k: int) -> int:

    hashmap = defaultdict(int)
    n = len(a)
    curr_sum = 0
    max_count = 0

    hashmap[0] = 1

    for i in range(n):

        curr_sum = curr_sum ^ a[i]

        max_count += hashmap[curr_sum^k]

        hashmap[curr_sum] += 1

    return max_count

a = [4, 2, 2, 6, 4]
k = 6
print(subarraysWithSumK(a=a, k=6))


a = [5, 6, 7, 8, 9]
k = 5
print(subarraysWithSumK(a=a, k=6))
```

## 1. Longest Common Prefix.

```python
def longestCommonPrefix(strs: List[str]) -> str:

    '''
    -> Common prefix

        Brute force approach.

        - sort the strs with respect to the length
        - take the smalles element as reference as this can be the max ans we can accept.
        - for each element in smallest element check.
    '''

    strs = sorted(strs, key=len)

    element = strs[0]
    ptr = 0
    ans = ""

    for e in element:

        flag = True

        for i in strs[1:]:
            if e != i[ptr]:
                flag = False
                break

        if not flag: break

        ans += e
        ptr += 1

    return ans


###method 2
def longestCommonPrefix(self, strs: List[str]) -> str:

    if not strs:
        return ""

    prefix = ""
    # iterate through characters of each string simultaneously
    for chars in zip(*strs):
        # if each letter is the same
        #print(chars,set(chars))
        if len(set(chars)) == 1:
            # add the letter to the prefix
            prefix += chars[0]
        else:
            # we found a mismatch, so return the longest prefix we found
            return prefix

    return prefix


```

```python
def countSubStrings(s: str, k: int) -> int:

    ###brute force, create all possiable substrings

    ### Time complixity : O(N^3)
    ### Space complixity : O(1)
    n = len(s)
    count = 0

    for i in range(n):
        item = s[i]
        for j in range(i+1,n):
            item += s[j]
            if len(set(item)) == k:
                count += 1

    ###also need to add len of string in case k == 1
    if k == 1:
        count += n

    return count

    ### Reduce the approach to O(N^2) by making the inner checking == O(1) by using more space.
    ### can use hashmap of type dict or list

    ###Time Complexity: O(N^2)
    ###Space Complexity: O(26) -> O(1)

    n = len(s)
    count = 0

    for i in range(n):

        ### for each subarray needs a new hashmap
        hashmap = [0] * 26
        unique_freq = 0

        for j in range(i, n):

            ###calcualte idx, 97 is acii for a
            idx = 97 - ord(s[j])

            ###now use the hashmap to think of
            if hashmap[idx] == 0:
                unique_freq += 1

            hashmap[idx] += 1

            if unique_freq == k:
                count += 1
            elif unique_freq > k:
                break

    return count
```

## 4. Count All Palindromic Subsequence in a given String

Input : str = "abcd"

Output : 4

Explanation :- palindromic  subsequence are : "a" ,"b", "c" ,"d" 

Input : str = "aab"

Output : 4

Explanation :- palindromic subsequence are :"a", "a", "b", "aa"

Input : str = "aaaa"

Output : 15

```python
def combinationSum(candidates: List[int], target: int) -> List[List[int]]:
    
    ### Method 1
    out = []
    n = len(candidates)

    def worker(idx, sum_, ans):
        nonlocal out

        if sum_ > target: return

        if sum_ == target: 
            # ans.sort()
            # if ans not in out:
                # out.append(ans)
            out.append(ans)
            return 
        
        for i in range(idx, n):
            worker(i, sum_+candidates[i], ans+[candidates[i]])
    
    worker(0, 0, [])

    return out

    ### Method 2
    out = []

    def worker(idx, target_, ans):
        nonlocal out

        # if target < 0: return

        if target_ == 0: 
            out.append(ans.copy())
            return
        
        if idx == len(candidates): return 

        ###not pick
        worker(idx+1, target_, ans)

        ###pick
        if candidates[idx] <= target_:
            worker(idx, target_-candidates[idx], ans + [candidates[idx]])

    
    worker(0, target, [])

    return out
```

## Combination Sum II

```python
def combinationSum2(candidates: List[int], target: int) -> List[List[int]]:

    candidates.sort()

    out = []
    
    n = len(candidates)

    def worker(idx, sum_, ans):
        nonlocal out

        if sum_ > target: return

        if sum_ == target: 
            out.append(ans)
            return 
        
        for i in range(idx, n):
            if i > idx and candidates[i] == candidates[i-1]: continue
            worker(i+1, sum_+candidates[i], ans+[candidates[i]])
    
    worker(0, 0, [])

    return out
```
