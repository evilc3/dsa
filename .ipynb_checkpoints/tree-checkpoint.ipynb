```python
class Node:

    def __init__(self,val, right, left):
        self.val = val 
        self.right = right 
        self.left = left

class Tree:

    '''
    Creates binary tree.
            2
        3      4
     5    6  7   8

    node_list = [2, 3, 4, 5, 6, 7, 8]

           2
        3     N
    
    node_list = [2, 3, N]
    '''

    def __init__(self, node_list):
            
        # left_child = 2 * i + 1
        # right_child = 2 * (i + 1)
        self.root = self.build_tree(node_list=node_list)


    def build_tree(self, node_list):

        def worker(idx):

            if idx >= len(node_list):
                return None 
            if node_list[idx] is None:
                return None 
                
            node = Node(val = node_list[idx], left = None, right = None)

            node.left = worker(2 * idx + 1)
            node.right = worker(2 * (idx+1))

            return node 
        
        return worker(0)

    
    def pre_order(self):
        ## root-left-right
        def worker(node):

            if not node: return 

            print(node.val, end = ' -> ')
            worker(node.left)
            worker(node.right)

        worker(self.root)
        print()
    
    def inorder(self):
        ## left-root-right
        def worker(node):

            if not node: return 

            worker(node.left)
            print(node.val, end = ' -> ')
            worker(node.right)
        
        worker(self.root)
        print()
    
    def post_order(self):
        ## left-right-root
        def worker(node):

            if not node: return 

            worker(node.left)
            worker(node.right)
            print(node.val, end = ' -> ')
        
        worker(self.root)
        print()
    
    def level_order(self):

        q = [self.root]

        while q != []:
            
            node = q.pop(0)
            print(node.val)
             
            if node.left:
                q.append(node.left)
            if node.right:
                q.append(node.right)
                        
        print()
    
    def level_order_2(self):

        q = [self.root]

        level = 0
        level_dict = {}

        while q != []:
            
            for i in range(len(q)):
                node = q.pop(0)
                
                if level not in level_dict:
                    level_dict[level] = [node.val]
                else:
                    level_dict[level].append(node.val)

                if node.left:
                    q.append(node.left)
                if node.right:
                    q.append(node.right)
            
            print(f'Level : {level} Nodes : {level_dict[level]}')
            level += 1
            
        print()
    
    def level_order_3(self):
        '''
        Same as level_order_3 but uses seperator
        ''' 

        q = [self.root, None]
        

        while q != []:
            
            node = q.pop(0)
            
            if not node:
                print()
                if q != []:
                    q.append(None)
            else:
                print(node.val, end = ' -> ')
             
                if node.left:
                    q.append(node.left)
                if node.right:
                    q.append(node.right)
                        
        print()
    
    def height(self):
        ### depth = height -1
        def worker(node):
            if not node:
                return 0
            return 1 + max(worker(node.left), worker(node.right))

        print(f'Height of tree : {worker(self.root)}')
    
    def diameter(self):
        ### O(N^2)
        def height(node):
            if not node:
                return 0
            return 1 + max(worker(node.left), worker(node.right))
        
        def worker(node):

            if not node:
                return 0 
            
            left = worker(node.left)
            right = worker(node.right)
            root = height(node.left) + height(node.right) + 1

            return max(root, left, right)

        return worker(self.root)
    
    def diameter_fast(self):
        ### Code by love babar.
        ## O(N) ***
        def worker(node):
            if not node:
                return 0,0
            
            dl, hl = worker(node.left)
            dr, hr = worker(node.right)

            op1 = dl 
            op2 = dr 
            op3 = hl + hr + 1

            d = max(op1,op2,op3)
            h = max(hr,hl) + 1
            return d, h

        return worker(self.root)[0]

        

node_list = [2, 3, 4, 5, 6, 7, 8]
tree = Tree(node_list)
```

```python
### check for balance tree
# https://practice.geeksforgeeks.org/problems/check-for-balanced-tree/1
def height(node):

    if not node: return 0 
    return 1 + max(height(node.left), height(node.right))

### TLE
def is_balance_tree(root):
    # TC : N^2
    if not root: return True 

    op1 = is_balance_tree(root.right)
    op2  = is_balance_tree(root.left)
    op3 = abs(height(root.left) - height(root.right)) <=1

    return op1 and op2 and op3

def is_balance_tree_fast(root):
    # TC : N
    if not root: return (True, 0)

    op1, lh = is_balance_tree_fast(root.right)
    op2, rh = is_balance_tree_fast(root.left)
    op3 = abs(lh - rh) <= 1

    x = op1 and op2 and op3
    y = 1 + max(lh,rh)
    return (x, y)


node_list1 = [1,10,39,5,None]
tree1 = Tree(node_list1)
print(is_balance_tree_fast(tree1.root))

node_list1 = [1,10,None,5,None]
tree1 = Tree(node_list1)
print(is_balance_tree_fast(tree1.root))


```

```python
### Morris Traversals
### PreOrder 
### Helps reduce space complixity to O(1) constant by creatime tmp linls
'''
- Set a current : Initially my root node.
- if current of left is null print current and set current to its right child
else
- if current of left is not null then 
    - find the predecessor: 
        - go left once.
        - keep going right until child is None
    - if predecessor.right is Null 
      - SET it to current 
      - SET current to current.left
    else
    - if predecessor.right is not null
      - set predecessor.right == null 
      - print current 
      - set current -> right
'''


def morris_traversal(root):

    ans = []

    current = root 

    while current:

        if not current.left:
            ans.append(current.data)
            ### for preorder
            # current = current.left 
            ### for inorder
            current = current.right
            

        else:

            predecessor = current.left
            
            while predecessor.right is not None and predecessor.right != current:
                predecessor = predecessor.right

            if not predecessor.right:
                predecessor.right = current 
                current = current.left 
            else:
                predecessor.right = None
                ans.append(current.data)
                current = current.right


    return ans

```

```python
# https://leetcode.com/problems/binary-tree-inorder-traversal/submissions/
class Solution:
    
    def morris_traversal(self, root):

        ans = []

        current = root 

        while current:

            if not current.left:
                ans.append(current.val)
                current = current.right 

            else:

                predecessor = current.left

                while predecessor.right is not None and predecessor.right != current:
                    predecessor = predecessor.right

                if not predecessor.right:
                    predecessor.right = current 
                    current = current.left 
                else:
                    predecessor.right = None
                    ans.append(current.val)
                    current = current.right


        return ans
    
    def inorderTraversal(self, root: Optional[TreeNode]) -> List[int]:
        
        return self.morris_traversal(root)
```

```python
### ZigZag Traverlsal/ spiral traversal 

### use level order get all the levels then keep a bool to inverst printing order.

def zigzag(self):

    q = [self.root]

    level = 0
    level_dict = {}

    while q != []:
        
        for i in range(len(q)):
            node = q.pop(0)
            
            if level not in level_dict:
                level_dict[level] = [node.val]
            else:
                level_dict[level].append(node.val)

            if node.left:
                q.append(node.left)
            if node.right:
                q.append(node.right)
        
        print(f'Level : {level} Nodes : {level_dict[level]}')
        level += 1
        
    print()
```

```python
class Solution:
    def verticalTraversal(self, root: Optional[TreeNode]) -> List[List[int]]:
        
        '''
        vertical order traveral works on axis 
        
        for root axis == 0
        when going to left axis - 1
        when goinf to right axis + 1 , 
        also need a level 
        when increase every node
        '''
        
        q = [(root, 0, 0)]
        mapp = {}
        axis_mini = float('inf')
        axis_maxi = float('-inf')
        
        while q != []:
            
            node, axis, level = q.pop(0)
            
        
            if axis not in mapp:
                mapp[axis] = [(level,node.val)]
            else:
                mapp[axis] += [(level, node.val)]
            
            axis_mini = min(axis_mini, axis)
            axis_maxi = max(axis_maxi, axis)
            
            if node.left:
                q.append((node.left,axis-1,level+1))
            
            if node.right:
                q.append((node.right,axis+1,level+1))
        
        # print(mapp)
        
        out = []
        
        for i in range(axis_mini, axis_maxi + 1):
            tmp = []
            for j in sorted(mapp[i]):
                tmp.append(j[1])
            out.append(tmp)
        
        return out
```

```python
### LCA Binary Tree 

def lca(root, n1, n2):

    def worker(node):

        if not node: return None

        if node.data == n1 or node.data == n2:
            return node.data
        
        left = worker(node.left)
        right = worker(node.right)

        if not left and not right:
            return None

        if left and not right:
            return left 
        
        if not left and right:
            return right
        
        return node 

    return worker(root)
```

```python
#### Two Sum in BST 
### apply inorder 
### solve like to sum using two pointer as the array is not sorted.
# https://www.codingninjas.com/codestudio/problems/two-sum-in-a-bst_1062631?leftPanelTab=1


def twoSumInBST(root, target):
    # Write your code here
    
     out = []
    
     def inorder(root):
        
        if not root: return None 
        
        inorder(root.left)
        out.append(root.data)
        inorder(root.right)
        
     inorder(root)
    
     p1 = 0
     p2 = len(out) - 1
    
     while p1 < p2:
           sum = out[p1] + out[p2] 
           if sum == target:
               return True 
           elif sum > target:
               p2 -= 1
           else:
                p1 += 1
                
     return False
```

```python
## Normal BST to Balanced BST 
class TreeNode:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

    def __del__(self):
        if self.left:
            del self.left
        if self.right:
            del self.right



def binaryTreeToBst(root):
	# Write your code here.
	
    out = []
    
    def inorder(root):
        
        if not root: return None 
        
        inorder(root.left)
        out.append(root)
        inorder(root.right)
        
    inorder(root)
    
    
    def build(l,h):
        
        if l > h: return None
        
        
        mid = (h+l) // 2
        
        node = TreeNode(data = out[mid])
        
        node.left = build(l,mid-1)
        node.right = build(mid+1,h)
        
        return node 
    
    return build(0, len(out) - 1)
```
