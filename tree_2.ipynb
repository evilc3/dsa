```python
class Node:

    def __init__(self,val, right, left):
        self.val = val 
        self.right = right 
        self.left = left

class Tree:

    '''
    Creates binary tree.
            2
        3      4
     5    6  7   8

    node_list = [2, 3, 4, 5, 6, 7, 8]

           2
        3     N
    
    node_list = [2, 3, N]
    '''

    def __init__(self, node_list):
            
        # left_child = 2 * i + 1
        # right_child = 2 * (i + 1)
        self.root = self.build_tree(node_list=node_list)


    def build_tree(self, node_list):

        def worker(idx):

            if idx >= len(node_list):
                return None 
            if node_list[idx] is None:
                return None 
                
            node = Node(val = node_list[idx], left = None, right = None)

            node.left = worker(2 * idx + 1)
            node.right = worker(2 * (idx+1))

            return node 
        
        return worker(0)

    
    def pre_order(self):
        ## root-left-right
        def worker(node):

            if not node: return 

            print(node.val, end = ' -> ')
            worker(node.left)
            worker(node.right)

        worker(self.root)
        print()
    
    def inorder(self):
        ## left-root-right
        def worker(node):

            if not node: return 

            worker(node.left)
            print(node.val, end = ' -> ')
            worker(node.right)
        
        worker(self.root)
        print()
    
    def post_order(self):
        ## left-right-root
        def worker(node):

            if not node: return 

            worker(node.left)
            worker(node.right)
            print(node.val, end = ' -> ')
        
        worker(self.root)
        print()
    
    def level_order(self):

        q = [self.root]

        while q != []:
            
            node = q.pop(0)
            print(node.val)
             
            if node.left:
                q.append(node.left)
            if node.right:
                q.append(node.right)
                        
        print()
    
    def level_order_2(self):

        q = [self.root]

        level = 0
        level_dict = {}

        while q != []:
            
            for i in range(len(q)):
                node = q.pop(0)
                
                if level not in level_dict:
                    level_dict[level] = [node.val]
                else:
                    level_dict[level].append(node.val)

                if node.left:
                    q.append(node.left)
                if node.right:
                    q.append(node.right)
            
            print(f'Level : {level} Nodes : {level_dict[level]}')
            level += 1
            
        print()
    
    def level_order_3(self):
        '''
        Same as level_order_3 but uses seperator
        ''' 

        q = [self.root, None]
        

        while q != []:
            
            node = q.pop(0)
            
            if not node:
                print()
                if q != []:
                    q.append(None)
            else:
                print(node.val, end = ' -> ')
             
                if node.left:
                    q.append(node.left)
                if node.right:
                    q.append(node.right)
                        
        print()
    
    def height(self):
        ### depth = height -1
        def worker(node):
            if not node:
                return 0
            return 1 + max(worker(node.left), worker(node.right))

        print(f'Height of tree : {worker(self.root)}')
    
    def diameter(self):
        ### O(N^2)
        def height(node):
            if not node:
                return 0
            return 1 + max(worker(node.left), worker(node.right))
        
        def worker(node):

            if not node:
                return 0 
            
            left = worker(node.left)
            right = worker(node.right)
            root = height(node.left) + height(node.right) + 1

            return max(root, left, right)

        return worker(self.root)
    
    def diameter_fast(self):
        ### Code by love babar.
        ## O(N) ***
        def worker(node):
            if not node:
                return 0,0
            
            dl, hl = worker(node.left)
            dr, hr = worker(node.right)

            op1 = dl 
            op2 = dr 
            op3 = hl + hr + 1

            d = max(op1,op2,op3)
            h = max(hr,hl) + 1
            return d, h

        return worker(self.root)[0]

        

node_list = [2, 3, 4, 5, 6, 7, 8]
tree = Tree(node_list)
```

```python
### check for balance tree
# https://practice.geeksforgeeks.org/problems/check-for-balanced-tree/1
def height(node):

    if not node: return 0 
    return 1 + max(height(node.left), height(node.right))

### TLE
def is_balance_tree(root):
    # TC : N^2
    if not root: return True 

    op1 = is_balance_tree(root.right)
    op2  = is_balance_tree(root.left)
    op3 = abs(height(root.left) - height(root.right)) <=1

    return op1 and op2 and op3

def is_balance_tree_fast(root):
    # TC : N
    if not root: return (True, 0)

    op1, lh = is_balance_tree_fast(root.right)
    op2, rh = is_balance_tree_fast(root.left)
    op3 = abs(lh - rh) <= 1

    x = op1 and op2 and op3
    y = 1 + max(lh,rh)
    return (x, y)


node_list1 = [1,10,39,5,None]
tree1 = Tree(node_list1)
print(is_balance_tree_fast(tree1.root))

node_list1 = [1,10,None,5,None]
tree1 = Tree(node_list1)
print(is_balance_tree_fast(tree1.root))


```

```python
### Morris Traversals
### PreOrder 
### Helps reduce space complixity to O(1) constant by creatime tmp linls
'''
- Set a current : Initially my root node.
- if current of left is null print current and set current to its right child
else
- if current of left is not null then 
    - find the predecessor: 
        - go left once.
        - keep going right until child is None
    - if predecessor.right is Null 
      - SET it to current 
      - SET current to current.left
    else
    - if predecessor.right is not null
      - set predecessor.right == null 
      - print current 
      - set current -> right
'''


def morris_traversal(root):

    ans = []

    current = root 

    while current:

        if not current.left:
            ans.append(current.data)
            ### for preorder
            # current = current.left 
            ### for inorder
            current = current.right
            

        else:

            predecessor = current.left
            
            while predecessor.right is not None and predecessor.right != current:
                predecessor = predecessor.right

            if not predecessor.right:
                predecessor.right = current 
                current = current.left 
            else:
                predecessor.right = None
                ans.append(current.data)
                current = current.right


    return ans

```

```python
# https://leetcode.com/problems/binary-tree-inorder-traversal/submissions/
class Solution:
    
    def morris_traversal(self, root):

        ans = []

        current = root 

        while current:

            if not current.left:
                ans.append(current.val)
                current = current.right 

            else:

                predecessor = current.left

                while predecessor.right is not None and predecessor.right != current:
                    predecessor = predecessor.right

                if not predecessor.right:
                    predecessor.right = current 
                    current = current.left 
                else:
                    predecessor.right = None
                    ans.append(current.val)
                    current = current.right


        return ans
    
    def inorderTraversal(self, root: Optional[TreeNode]) -> List[int]:
        
        return self.morris_traversal(root)
```

```python
### ZigZag Traverlsal/ spiral traversal 

### use level order get all the levels then keep a bool to inverst printing order.

def zigzag(self):

    q = [self.root]

    level = 0
    level_dict = {}

    while q != []:
        
        for i in range(len(q)):
            node = q.pop(0)
            
            if level not in level_dict:
                level_dict[level] = [node.val]
            else:
                level_dict[level].append(node.val)

            if node.left:
                q.append(node.left)
            if node.right:
                q.append(node.right)
        
        print(f'Level : {level} Nodes : {level_dict[level]}')
        level += 1
        
    print()
```

```python
class Solution:
    def verticalTraversal(self, root: Optional[TreeNode]) -> List[List[int]]:
        
        '''
        vertical order traveral works on axis 
        
        for root axis == 0
        when going to left axis - 1
        when goinf to right axis + 1 , 
        also need a level 
        when increase every node
        '''
        
        q = [(root, 0, 0)]
        mapp = {}
        axis_mini = float('inf')
        axis_maxi = float('-inf')
        
        while q != []:
            
            node, axis, level = q.pop(0)
            
        
            if axis not in mapp:
                mapp[axis] = [(level,node.val)]
            else:
                mapp[axis] += [(level, node.val)]
            
            axis_mini = min(axis_mini, axis)
            axis_maxi = max(axis_maxi, axis)
            
            if node.left:
                q.append((node.left,axis-1,level+1))
            
            if node.right:
                q.append((node.right,axis+1,level+1))
        
        # print(mapp)
        
        out = []
        
        for i in range(axis_mini, axis_maxi + 1):
            tmp = []
            for j in sorted(mapp[i]):
                tmp.append(j[1])
            out.append(tmp)
        
        return out
```

```python
# https://leetcode.com/problems/maximum-product-of-splitted-binary-tree/submissions/
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def maxProduct(self, root: Optional[TreeNode]) -> int:
        
        ## LRN
        
        '''
        1. Get total sum 
        2. Iterate throught all possiable solutiions 
        
                1
            2      3 
            
            choise 1 : (1 + 2) * 3   ignore edge between 1 - 3
            choise 2 : (1 + 3) * 2   ignore edge between 1 - 2
            
            >> Simplity 
            
            >> total_sum = 1 + 2 + 3 = 6 
            
            >>> choise 1 : (6 - 3) * 3 = 9
            >>> choise 2 : (6 - 2) * 2 = 8
            
            >> Traversal Order is Post-Order
            
            
            Solving for example : 

            -----------------------------------------
            
                         1
                    2         3
                4      5   6


            total_sum = 1 + 2 + 3 + 4 + 5 + 6 = 21
            
            for node 4 :  (21 - 4) * 4 =  17 * 4 =  68
            for node 5 :  (21 - 5) * 5 =  16 * 5 =  80
            (need the num of the entire subtree.) hence I need to return.
            for node 2 :  (21 - (2 + 5 + 4)) * 11 =   10 * 11 = 110
            
            for node 6 : (21 - 6) * 6 :   15 * 6 =  90
            for node 3 : (21 - (3 + 6)) * 9 : 12 * 9 =  108
            
            
            (cant calculate node before the right subtree)
            
            for node: 
                
                choise 1 :
                        for node 2 :  (21 - (2 + 5 + 4)) * 11 =   10 * 11 = 110
                choise 2 : 
                       for node 3 : (21 - (3 + 6)) * 9 : 12 * 9 =  108
                
                take max of choise 1 and chose 2 and store in a GLOBAL VARIABLE
        '''
        
        def get_total_sum(root):
            
            if not root:
                return 0 
            
            return root.val + get_total_sum(root.left) + get_total_sum(root.right)
        
        total_sum = get_total_sum(root)
        
        res = 0
        
        def get_max_product(root):
            nonlocal res, total_sum
            
            if not root: return 0 
            
            left = get_max_product(root.left)
            right = get_max_product(root.right)
            
            res = max(res, left * (total_sum - left), right * (total_sum - right))
            
            ### sum of subtree
            
            return root.val + left + right 
        
        get_max_product(root)
        
        return res % (10 ** 9 + 7)
            
```

```python
# https://leetcode.com/problems/reverse-odd-levels-of-binary-tree/submissions/
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def reverseOddLevels(self, root: Optional[TreeNode]) -> Optional[TreeNode]:
        
        '''
                    1        0
                2      3     1
              4   5  6   7   2 
              
              dfs(root = root, pair_node=root, level = 0)
              
                  if not root: return None

                  left = dfs(root.left, pair_node.right, level + 1)
                  right = dfs(root.right, pair_node.left, level + 1)

                  if level is odd: 
                      n = Node(pair_node.val)

                      n.left = left 
                      n.right = right

                      return node
                      
                  return root
                  
        root1, root2 
        (root1.left,root1.right), (root1.right, root2.left)
        
        
        
        '''
        
        def worker(root, pair_node, level):
            
            if not root or not pair_node: return None 
            
            if level % 2 != 0:
                root.val, pair_node.val = pair_node.val, root.val
            
            worker(root.left, pair_node.right, level + 1)
            worker(root.right, pair_node.left, level + 1)
            

        
        worker(root = root.left, pair_node = root.right, level = 1)
        
        return root                    
```

```python
# https://leetcode.com/problems/delete-leaves-with-a-given-value/submissions/
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def removeLeafNodes(self, root: Optional[TreeNode], target: int) -> Optional[TreeNode]:
        
        
        def worker(root):
        
            if not root: return None 
            
            if not (root.left or root.right): 
                if root.val == target: 
                    return None 
                else:
                    return root
                
            
            left = worker(root.left)
            right = worker(root.right)
            
            root.left = left 
            root.right = right 
            ### is this Node now a leaf
            if not (root.left or root.right):
                if root.val == target:
                    return None 
            
            return root
        
        return worker(root)
```

```python
# https://leetcode.com/problems/lowest-common-ancestor-of-deepest-leaves/
# https://leetcode.com/problems/smallest-subtree-with-all-the-deepest-nodes/
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def lcaDeepestLeaves(self, root: Optional[TreeNode]) -> Optional[TreeNode]:
        
        # last_nodes = None
        def get_deepest_level_nodes(root):
            # nonlocal last_nodes 
            
            q = [root]
            last_nodes = None
            
            while q != []:
                
                size = len(q)
                last_nodes = q.copy()
                # print('last_nodes : ', last_nodes)
                for i in range(size):
                    
                    node = q.pop(0)
                    
                    if node.left:
                        q.append(node.left)
                    
                    if node.right:
                        q.append(node.right)
            return last_nodes
        
        def lca(root, node_a, node_b):
            
            if not root: return None
            
            if root.val == node_a or root.val == node_b:
                return root
            
            left = lca(root.left, node_a, node_b)
            right = lca(root.right, node_a, node_b)
            
            if not left and not right: return None 
            if not left and right: return right 
            if not right and left: return left 
            
            return root
                
        edge_nodes = get_deepest_level_nodes(root)
                
        ### edge case 
        if len(edge_nodes) == 1: return edge_nodes[0]
        
        node_a = edge_nodes[0].val
        node_b = edge_nodes[-1].val
    
        return lca(root, node_a, node_b)
```

```python
# https://leetcode.com/problems/smallest-subtree-with-all-the-deepest-nodes/
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def subtreeWithAllDeepest(self, root: TreeNode) -> TreeNode:
        
        def worker(root):
        
            if not root: return 0, None 
            
            left_d, left_lca = worker(root.left)
            right_d, right_lca = worker(root.right)
            
            if left_d == right_d:
                return left_d + 1, root
            
            if left_d > right_d:
                return left_d + 1, left_lca
            
            return right_d + 1, right_lca
        
        return worker(root)[1]
```
